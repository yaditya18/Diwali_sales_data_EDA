```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
df  = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
#drop or remove unwanted cloumns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
#remove null values
df.dropna(inplace=True)
```

```python
 #change data type
df['Amount'] = df['Amount'].astype('int')
```

# Exploratory Data Analysis

# Gender

```python
#count of each gender
ax = sns.countplot(x = 'Gender',data = df)

#for labels over plot
for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph we can see that number of female buyers is more than male .




# Age

```python
sns.countplot(data=df,x ='Age Group')
```

```python
ax = sns.countplot(data=df,x ='Age Group',hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph we can see most of the buyer belongs to 26-35 age group

# Occupation and Product Category

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df , x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df , x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x  = 'Product_Category' , y = 'Amount')
```

Highest sale amount in Food

```python

# Sales Trend
sns.lineplot(x='Orders', y='Amount', data=df)
plt.title('Sales Trend')
plt.show()

```

```python
# Correlation Analysis
correlation_matrix = df.corr(numeric_only = [True])
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Customer Purchase Behavior
customer_orders = df['User_ID'].value_counts()
sns.histplot(customer_orders, bins=30, kde=True)
plt.title('Customer Purchase Behavior')
plt.xlabel('Number of Orders')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average Order Value
average_order_value = df['Amount'].mean()
sns.barplot(x=['Average Order Value'], y=[average_order_value])
plt.title('Average Order Value')
plt.show()
```

```python
# Zone-wise Sales
sns.boxplot(x='Zone', y='Amount', data=df)
plt.title('Zone-wise Sales')
plt.show()
```
